```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
def showImg(img, colored=False, showAxis=False):
    plt.figure(figsize=(20,20))
    if not showAxis:
        plt.axis("off")
    if (len(img.shape) == 3):
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    elif(colored):
        plt.imshow(img)
    else:
        plt.imshow(img, 'gray')
```

```python
img1 = cv2.imread('./data/tax1.jpg')
img2 = cv2.imread('./data/tax2.jpg')
img3 = cv2.imread('./data/tax3.jpg')

plt.axis("off")
fig, [ax1, ax2, ax3] = plt.subplots(1,3)
fig.set_size_inches((20,5))
ax1.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax3.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB));
```

```python
colors = np.array([
    [255,0,0],
    [0,255,0],
    [0,0,255],
    
    [0,255,255],
    [255,0,255],
    [255, 255, 0],
    
    [0, 0, 0],
    [255, 100, 100],
])
def showImgAndPoints(img, points):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    for i in range(len(points)):
        plt.plot(points[i,0], points[i,1], color=colors[i]/255, marker='x')
    plt.show();

points_1 = np.array([
    [270,210],
    [1062,120],
    [1844,200],
    [172,1258],
    [1954,1238],
    [994,318],
    [1130,318],
    [1062,436],
])


points_2 = np.array([
    [50,158],
    [283,43],
    [355,33],
    [47,426],
    [353,410],
    [233,133],
    [260,123],
    [249,164],
])


points_3 = np.array([
    [354,44],
    [405,105],
    [600,224],
    [351,431],
    [608,455],
    [428,172],
    [453,186],
    [437,212],
])
points_1 = np.int32(points_1)
points_2 = np.int32(points_2)
points_3 = np.int32(points_3)

showImgAndPoints(img1, points_1)
showImgAndPoints(img2, points_2)
showImgAndPoints(img3, points_3)
```

```python
def normalizePoints(points):
    points = np.append(points.T, [np.ones(len(points))], axis=0).T
    
    mean_x, mean_y, _ = np.mean(points, axis=0)
    
    diffs_x = points[:,0] - mean_x 
    diffs_y = points[:,1] - mean_y
    
    scale = np.sqrt(2) / np.mean(np.sqrt(diffs_x**2 + diffs_y**2))
    
    T = np.array([
        [scale, 0,     -(scale*mean_x)],
        [0,     scale, -(scale*mean_y)],
        [0,     0,     1],
    ])
    res_points = T @ points.T
    return res_points, T

def fundamentalMatrix(points1, points2):
    p1, T1 = normalizePoints(points1)
    p2, T2 = normalizePoints(points2)
    
    A = np.array([
        p2[0]*p1[0], 
        p2[0]*p1[1],
        p2[0],
        p2[1]*p1[0],
        p2[1]*p1[1],
        p2[1],
        p1[0],
        p1[1],
        np.ones(len(points1)),
    ])
    
    U, S, V = np.linalg.svd(A.T)
    
    min_V = V[-1]
    min_V = min_V.reshape(3,3)
    
    U, S, V = np.linalg.svd(min_V)
    F = T2.T @ U @ np.diag([S[0], S[1], 0]) @ V @ T1
    return F
```

```python
def showLines(img1, img2, lines, points1, points2):
    img1=img1.copy()
    img2=img2.copy()
    h, w, c = img1.shape
    
    for i in range(len(lines)):
        line = lines[i]
        a, b, c = line

        point1 = tuple (points1[i])
        point2 = tuple (points2[i])
        
        from_points = (0, int(-c / b ))
        to_points = (w, int(-(c + a * w)/ b))

        color = tuple ([int(x) for x in colors[i]])
        # draw line on 1st image
        img1 = cv2.line(img1, from_points, to_points, color, 3)
        # add points
        img1 = cv2.circle(img1, point1, 5, color, -1)
        img2 = cv2.circle(img2, point2, 5, color, -1)
        

    fig, [ax1, ax2] = plt.subplots(1,2)
    fig.set_size_inches((15,15))
    ax1.imshow(img1)
    ax2.imshow(img2)
    plt.show();
```

```python
def epilines(points1, points2):
    F_matrix = fundamentalMatrix(points1, points2)
    res = []
    for point_2 in points2:
        vector = np.array([point_2[0], point_2[1], 1])
        line = F_matrix.T @ vector
        res.append(line)

    return np.array(res)
```

```python
def compare2image(img1,img2,p1,p2):
    lines = epilines(p1, p2)
    showLines(img1, img2, lines, p1, p2)
```

```python
compare2image(img1,img2,points_1,points_2)
```

```python
compare2image(img2,img3,points_2,points_3)
```
